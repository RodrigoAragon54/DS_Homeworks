# 1.	INTRODUCCIÓN

## ¿Qué son los sistemas de control de versiones?


Los sistemas de control de versiones son sistemas que se utilizan para llevar un control de los cambios y actualizaciones generadas en documentación relacionada a un proyecto, estos a su vez son herramientas colaborativas y permiten llevar un registro histórico (bitácora) de todas las actualizaciones generadas a través del tiempo incluyendo información relevante para la trazabilidad de la información respondiendo ¿Qué cambio?, ¿Dónde cambio?, ¿Quién lo hizo? Y ¿Cuándo se hizó?

¿Cuando se utilizan los sistemas de control de versiones?

Estos sistemas pueden ser utilizados en distintos tipos de proyectos, desde proyectos personales hasta grandes y complejos proyectos con participación de distintos equipos de profesionales. 

¿Qué es Git y cual es su diferencia con Github y GitLab?

Git es un sistema de control de versionamiento mientras que Github y GitLab son servicios en línea que utilizan Git.

# 2.	CONCEPTOS CLAVE

## 2.1 Modelo de Datos en Git
Git utiliza el modelado jerárquico de directorios y archivos, y los captura como “fotografías”a través del tiempo. 
En la jerga de Git un archivo es llamado “blob” mientras que un directorio es llamado “árbol”, de manera común puede compararse como la estructura de documentos que conocemos como Carpeta—>Archivo

## 2.2 Grafo Acíclico y dirigido (GAD) y su uso en Git

### 2.2.1 ¿Qué es un Grafo Acíclico Dirigido?

Es un conjunto de vértices y aristas que no contienen ciclos, mientras que un grafo dirigido es un grafo en el que las aristas tienen una dirección asociada.
En un GAD, las aristas van de un vértice a otro vértice, pero no se puede regresar por el mismo camino para formar un ciclo. Esto hace que los DAG sean útiles para representar relaciones de dependencia y orden en una variedad de aplicaciones, incluyendo la programación de tareas, la optimización de trabajos, la modelación de procesos y el análisis de datos.


### 2.2.2 ¿Cómo se relaciona con Git?

En Git el concepto de GAD es utilizado para llevar a cabo el proceso de control de versiones de la documentación de un proyecto y para poder interactuar en Git con los sets de información que componen un proyecto en se hace uso de comandos, los cuales permiten realizar diferentes tareas de interacción y creación.  
En la jerga de Git un Commit hace referencia a la acción de guardar una versión de un set de documentación en un momento dado, como se menciono anteriormente una “fotografía”, es importante mencionar que esta acción contiene metadata importante para la trazabilidad de cambios como el responsable de los cambios, los cambios realizados, fecha, etc.
¿Pero dónde sucede un commit? Un commit es una acción que puede suceder en cualquier branch. ¿Pero qué es un Branch?  Un Branch es una rama separa de la línea de desarrollo de un proyecto

# 3. COMANDOS BASICOS DE GIT Y EJEMPLOS

* git init: Este comando se utiliza para inicializar un nuevo repositorio Git en la carpeta actual. Por ejemplo, si deseas crear un nuevo repositorio en una carpeta llamada "mi-proyecto”

* git clone: Este comando se utiliza para clonar un repositorio existente en tu computadora. Por ejemplo, si deseas clonar un repositorio llamado "mi-proyecto-remoto" 

* git add: Este comando se utiliza para agregar archivos a la zona de "staging" antes de hacer un commit. Por ejemplo, si deseas agregar todos los archivos modificados en tu proyecto a la zona de "staging"

* git commit: Este comando se utiliza para guardar los cambios en la zona de "staging" en el repositorio local. Por ejemplo, si deseas hacer un commit con el mensaje "Añadir nuevas funcionalidades"

* git push: Este comando se utiliza para enviar los cambios desde el repositorio local al repositorio remoto. Por ejemplo, si deseas enviar los cambios realizados en tu repositorio local al repositorio remoto llamado "origin"

* git pull: Este comando se utiliza para descargar los cambios realizados en el repositorio remoto y actualizar el repositorio local. Por ejemplo, si deseas descargar los cambios realizados en el repositorio remoto llamado "origin"
